```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/kaggle/input/financeal-data/Financeal_data.csv')
df.head()
```

```python
#replace all the empty cells by nan

df= df.replace(r'ˆ\s*$', np.nan, regex = True)
```

```python
sns.countplot( x = 'gender', data = df, linewidth = 3, palette = 'Set2', edgecolor= "black")
plt.show()
```

```python
sns.countplot(x = 'age', data = df, palette = 'Set3', linewidth = 2)
```

```python
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(x= df['gender'], hue= df['Investment_Avenues'], palette = 'summer', linewidth = 3, edgecolor = 'white')
plt.title("Investment Avenues")

plt.subplot(1,2,2)
sns.countplot(x= df['gender'], hue= df['Stock_Marktet'], palette = 'hot', linewidth = 3, edgecolor = 'white')
plt.title("Stock Market")

plt.show()
```

```python
sns.countplot(x= df['Factor'],  palette = 'coolwarm', linewidth = 2)
```

```python
plt.figure(figsize = (16,8))
sns.pointplot(x= 'Purpose', y= 'age', data=df, linestyles = '--', capsize = .3, color = 'green')
plt.show()

#shows median Age
```

```python
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(hue= df['gender'], x= df['Duration'], palette = 'viridis', linewidth = 2, edgecolor = 'black')
plt.title("Duration")

plt.subplot(1,2,2)
sns.countplot(hue= df['gender'], x= df['Invest_Monitor'], palette = 'seismic', linewidth = 2, edgecolor = 'black')
plt.title("Investment Monitoring")

plt.show()
```
